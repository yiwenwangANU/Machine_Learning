##Explore the data

```python
import zipfile

zip_ref = zipfile.ZipFile('titanic.zip')
zip_ref.extractall()
zip_ref.close()
```

```python
import pandas as pd

training_set = pd.read_csv('train.csv')
test_set = pd.read_csv('gender_submission.csv').join(pd.read_csv('test.csv').set_index('PassengerId'),
                                                     on='PassengerId') 
training_set
```

```python
test_set
```

```python
training_set.info()
```

```python
training_set.describe()
```

```python
# Drop PassengerId, Name, Ticket
training_set = training_set.drop(columns = ['PassengerId', 'Name', 'Ticket'])
test_set = test_set.drop(columns = ['PassengerId', 'Name', 'Ticket'])
```

```python
training_set, test_set
```

##Take care of missing data

```python
for col in training_set.columns.values:
  print(f"In {col}, there are {training_set[col].isnull().sum()} nan value")
```

```python
training_set = training_set.drop(columns = ['Cabin'])
test_set = test_set.drop(columns = ['Cabin'])
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

imputer = SimpleImputer(strategy='most_frequent', missing_values=np.nan)
imputer = imputer.fit(training_set)
training_set.iloc[:,:] = imputer.transform(training_set)
training_set
```

```python
for col in training_set.columns.values:
  print(f"In {col}, there are {training_set[col].isnull().sum()} nan value")
```

```python
imputer = SimpleImputer(strategy='most_frequent', missing_values=np.nan)
imputer = imputer.fit(test_set)
test_set.iloc[:,:] = imputer.transform(test_set)
```

```python
X_train = training_set.iloc[:, 1:].values
y_train = training_set.iloc[:, 0].values

X_test = test_set.iloc[:, 1:].values
y_test = test_set.iloc[:, 0].values
```

```python
X_train, X_test
```

##Take care of the categorical values

```python
# Label Encode Sex
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X_train[:, 1] = le.fit_transform(X_train[:, 1])
X_test[:, 1] = le.fit_transform(X_test[:, 1])
```

```python
X_train, X_train.shape
```

```python
# One hot encode Embarked
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [-1])], remainder='passthrough')
X_train = np.array(ct.fit_transform(X_train))
X_test = np.array(ct.fit_transform(X_test))
```

```python
print(X_test[:5, :])
```

##Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

##Training the Xgboost Classification model on the Training set

```python
from xgboost import XGBClassifier
classifier = XGBClassifier()
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

##Training the Random Forest Classification model on the Training set

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
