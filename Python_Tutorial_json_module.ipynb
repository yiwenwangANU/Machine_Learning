#Load json(string) into a python dictonary

```python
import json
data = json.loads(states_string)
```

```python
print(data)
```

```python
for state in data['states']:
  print(state)
```

#Put python dictionary back to json(string)

```python
for state in data["states"]:
  del state['area_codes']

for state in data["states"]:
  print(state)
```

```python
new_string = json.dumps(data, indent=2)
print(new_string)
```

#Put python dictionary back to json file

```python
with open('states.json', 'w') as f:
  json.dump(data, f, indent=2)
```

```python
with open('states.json', 'r') as f:
  new_data = json.load(f)
print(new_data)
```

#Another example

```python
import json
from urllib.request import urlopen

with urlopen("https://ip-ranges.amazonaws.com/ip-ranges.json") as response:
    source = response.read()
```

```python
data = json.loads(source)
with open('ip_ranges.json', 'w') as f:
  json.dump(data, f, indent=2)
```

```python
data.keys()
```

```python
ip_region = dict()
for prefixe in data["prefixes"]:
  ip_region[prefixe['ip_prefix']] = prefixe['region']
print(ip_region)
```
